# **Evaluación Phi-3.5-mini en Español**

```python
# Importar librerías
import pandas as pd
import torch
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM

# Cargar dataset desde Excel
file_path = "/content/BenchmarkSpanish.xlsx"
df = pd.read_excel(file_path)

# Configurar modelo
model_id = "microsoft/Phi-3.5-mini-instruct"

# Cargar modelo y tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    torch_dtype=torch.bfloat16
)

# Crear pipeline
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=512,
    temperature=0.01,
    top_p=0.9,
    repetition_penalty=1.1
)
```

# **Intento 1**

```python
# Iterar sobre el dataset y generar respuestas con prompt Zero-shot
results = []


for idx, question in zip(df["id"], df["question"]):
    print(f"🔹 Question {idx}: {question}")

    prompt = f"""<|user|>

{question}
<|end|>
<|assistant|>"""

    output = generator(prompt)[0]["generated_text"]
    response = output.replace(prompt, "").strip()

    print(f"🔸 Answer: {response}\n" + "-" * 80)

    results.append({
        "id": idx,
        "question": question,
        "answer": response
    })



# Guardar resultados en Excel
df_results = pd.DataFrame(results)
df_results.to_excel("/content/ResultadosSpanishPhi1.xlsx", index=False)
print("✅ Archivo guardado: ResultadosSpanishPhi1.xlsx")
```
